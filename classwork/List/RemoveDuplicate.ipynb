```python
# Write a program to create a list of any 15 number ask the user to input any other number search this element into the list and remove all the occurence from the list excluding the first occurences


# Create a list of 15 numbers (you can modify values)
numbers = [10, 20, 30, 40, 20, 50, 60, 20, 70, 80, 90, 20, 100, 110, 20]

x = int(input("Enter a number to search and remove duplicates: "))
first_found = False
i = 0
result = []
while(i < len(numbers)):
  if(numbers[i] == x):
    if(first_found == False):
      result.append(numbers[i])
      first_found = True
  else:
    result.append(numbers[i])
  i += 1

print(result)

```
