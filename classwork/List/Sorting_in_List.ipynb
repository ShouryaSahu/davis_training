```python
# Write a program to create a list of any ten numbers and sort them in descending order without using any library method for sorting

# Create a list of 10 numbers
numbers = [12, 5, 8, 19, 3, 15, 7, 1, 10, 6]

# Sorting in descending order
n = len(numbers)

for i in range(n):
    for j in range(0, n - i - 1):
        if numbers[j] < numbers[j + 1]:  # Swap for descending order
            # Swap the elements
            temp = numbers[j]
            numbers[j] = numbers[j + 1]
            numbers[j + 1] = temp

# Print the sorted list
print("Sorted list in descending order:", numbers)
```
