```python
# Function to calculate simple interest

def simple_interest(p, r, t):
    # Formula: (P * R * T) / 100
    si = (p * r * t) / 100
    return si

# Taking input
p = float(input("Enter Principal: "))
r = float(input("Enter Rate: "))
t = float(input("Enter Time: "))

# Calling function
result = simple_interest(p, r, t)

print("Simple Interest:", result)
```
